```python
cols_to_drop = []

# loop over the columns in the DataFrame
for col in credit_df.columns:
    # check if the number of unique values in the column is equal to 1
    if credit_df[col].nunique() == 1 or credit_df[col].nunique() ==0:
        # if so, append the name of the column to the list of columns to drop
        cols_to_drop.append(col)

# drop the columns in the list of columns to drop
credit_df_drop = credit_df.drop(cols_to_drop, axis=1)
```
